#Mendapatkan semua image path

```python
from glob import glob
```

```python
fname = glob('drive/MyDrive/data_image/*/*.png')
fname
```

#X = Path -> Image -> Flatten

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([plt.imread(f).flatten() for f in fname])
X
```

```python
X.shape
```

#y = Path -> split -> ambil[3]

```python
y = np.array([int(f.split('/')[3]) for f in fname])
y
```

```python
y.shape
```

#Join sebagai 1 DataFrame

```python
df = pd.DataFrame(X, columns = range(1,785))
df.head()
```

```python
df['label'] = y
df.head()
```

#Save to csv

```python
df.to_csv('metadata_exercise', index=None)
```
